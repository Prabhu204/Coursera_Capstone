# 1. Preprocessing data from the wikipage for German Cities

```python
import wikipedia
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import unicodedata
```

```python
wiki = wikipedia.page("List of cities in Germany by population").html()
soup = BeautifulSoup(wiki)
soup
```

### Extract table html using BeautifulSoup

```python
table = soup.find("table")
table
```

```python
columns = [th.get_text().rstrip() for th in table.find("tr").find_all("th")]
```

```python
columns
```

```python
dataset = []
for row in table.find_all("tr")[1:]:
    data = dict(zip(columns, (unicodedata.normalize("NFKD", td.get_text()).strip() for td in row.find_all("td"))))
    data["Location"]= data["Location"].split("/")[2].split()
    data["Lat"] = data["Location"][0].split(";")[0]
#     data["Lat"] = data["Location"][0].split()
    data["Lan"] = unicodedata.normalize("NFKD", data["Location"][1]).strip()
    print(data["Lan"])
    dataset.append(data)
print(dataset)
```

### Dataframe will consist of three columns: PostalCode, Borough, and Neighborhood and its values

```python
df = pd.DataFrame(dataset)
df = df[["City","State","2015populationdensity","Lat", "Lan"]]
print(df.shape)
print("**********************")
df.head()
```

```python
df["Persqkmpopdensity"], df["Persqmipopdensity"] = df["2015populationdensity"].str.split("/km2",1).str
df["Persqmipopdensity"] = df["Persqmipopdensity"].str.split("/").str[0]

cities = df.State.unique().tolist()
cities
```

```python
df
```

### Conversion of dataframe object columns into digits

```python
import re
# df["Persqkm"] = df["Persqkm"].apply(lambda x: re.search(r'\d+', x).group())
df["Persqkmpopdensity"]= df.Persqkmpopdensity.str.split("").apply(lambda x: ''.join([i for i in x if i.isdigit()])).astype(int)
df["Persqmipopdensity"]= df.Persqmipopdensity.str.split("").apply(lambda x: ''.join([i for i in x if i.isdigit()])).astype(int)
print(df.dtypes)
```

```python
df.Lan.str.split("")
```

```python
df["Lan"]=df.Lan.str.split("").apply(lambda x: "".join(x[1:-2])).astype(float)
```

```python
df["Lat"] = df["Lat"].astype(float)
df["City"] = df["City"].astype("category")
df["State"] = df["State"].astype("category")
print(df.dtypes)
df
```

### Combing cities with a comma separated which are having same state.

```python
combi_cities = df.groupby(["State"])["City"].agg([('City', ', '.join)])
combi_cities.reset_index(inplace = True)
combi_cities
```

## Explore Cities in North Rhine-Westphalia

```python
LIMIT = 300
radius = 3000 # define radius


def getNearbyVenues(names, latitudes, longitudes, radius=radius):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

##### Now write the code to run the above function on each City and create a new dataframe called NRW_venues.¶

```python
NRW_venues = getNearbyVenues(names=df_NRW['City'],
                                   latitudes=df_NRW['Lat'],
                                   longitudes=df_NRW['Lan'])
```

## Analyze Each Neighborhood

```python
# one hot encoding
NRW_onehot = pd.get_dummies(NRW_venues[['Venue Category']], prefix="", prefix_sep="")
NRW_onehot = NRW_onehot.rename(columns = {"Neighborhood":"Neighborhood_venue" })
# add neighborhood column back to dataframe
NRW_onehot = pd.concat([NRW_venues["Neighborhood"], NRW_onehot], axis=1)
NRW_onehot.head()
```

```python
# And let's examine the new dataframe size.
NRW_onehot.shape
```

##### Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
NRW_grouped = NRW_onehot.groupby('Neighborhood').mean().reset_index()
NRW_grouped.head()
```

#### Let's confirm the new size

```python
NRW_grouped.shape
```

#### Let's print each neighborhood along with the top 10most common venues

```python
num_top_venues = 10

for hood in NRW_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = NRW_grouped[NRW_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

#### Let's put that into a pandas dataframe

#### First, let's write a function to sort the venues in descending order.


```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
#   print(row_categories)
    row_categories_sorted = row_categories.sort_values(ascending=False)
    return row_categories_sorted.index.values[0:num_top_venues]
```

##### Now let's create the new dataframe and display the top 10 venues for each neighborhood.

```python
num_top_venues = 40
indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = NRW_grouped['Neighborhood']

for ind in np.arange(NRW_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(NRW_grouped.iloc[ind, :], num_top_venues)

# print(neighborhoods_venues_sorted)

neighborhoods_venues_sorted.head()
```

## Cluster Neighborhoods

```python
from sklearn.cluster import KMeans

kclusters = 6

NRW_grouped_clustering = NRW_grouped.drop('Neighborhood', 1)
# print(NRW_grouped_clustering)
# run k-means clustering
kmeans = KMeans(n_clusters=kclusters)
kmeans.fit(NRW_grouped_clustering)

# check cluster labels generated for each row in the dataframe
print(kmeans.labels_)
len(kmeans.labels_)
```

##### Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

```python
# add clustering labels
# neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
neighborhoods_venues_sorted['Cluster Labels'] = kmeans.labels_
NRW_merged = df_NRW
# print(NRW_merged.head())

# merge NRW_grouped with NRW_data to add latitude/longitude for each neighborhood
# print(neighborhoods_venues_sorted.head())
NRW_merged = NRW_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='City')
NRW_merged = NRW_merged.iloc[:-1,:]
# NRW_merged['Cluster Labels'] = neighborhoods_venues_sorted['Cluster Labels'].astype(int)
NRW_merged.head()# check the last columns!
```

### Let's visualize the resulting clusters

```python
import numpy as np
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=8)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(NRW_merged['Lat'], NRW_merged['Lan'], NRW_merged['City'], NRW_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
